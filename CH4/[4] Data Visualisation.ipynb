# Data Visualisation

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

bmi_data = pd.read_csv("./Data/500_Person_Gender_Height_Weight_Index.csv")
print(bmi_data.head())

stock_data = pd.read_csv("./Data/AABA_2006-01-01_to_2018-01-01.csv")
print(stock_data.head())
```

## Histograms

```python
bmi_data["Height"].hist(bins=20)
plt.show()
```

## Bar Plots

```python
avg_weight_by_gender = bmi_data.groupby("Gender")["Weight"].mean()
print(avg_weight_by_gender)
```

```python
avg_weight_by_gender.plot(kind="bar", rot=0)
plt.show()
```

## Line plots

- good for plotting the change in a numeric variable over time

```python
stock_data["Date"]= pd.to_datetime(stock_data["Date"])
stock_2010 = stock_data.set_index("Date")
stock_2010 = stock_2010.sort_index().loc["2010":"2010"].reset_index()

stock_2010.plot(x="Date", y="Close", kind="line")
plt.show()
```

## Scatter plots

- good for seeing relationships between 2 numeric variables

```python
male_srt = bmi_data[bmi_data["Gender"] == "Male"].sort_values("Height")
female_srt = bmi_data[bmi_data["Gender"] == "Female"].sort_values("Height")

plt.scatter(x=male_srt["Height"], y=male_srt["Weight"], alpha=0.5)
plt.scatter(x=female_srt["Height"], y=female_srt["Weight"], alpha=0.5, c="orange")
plt.legend(["M", "F"])
plt.show()
```

## Missing Values

### 1. Finding Missing Values

- in a pandas data frame, a missing value is represented by NaN (Not a Number)
- .isna() gives us a boolean of whether a cell is missing a value
- .any() tells us if any of the values in the columns are true
- using .isna() and .any() together lets us see if there are missing values in any of the columns
- .sum() works similarly to .any() in that it gives the number of rows with missing data for that column

### 2. Dropping Missing Values

- not ideal for datasets with a lot of missing values 
- note that missing values might be due to some quirk of the data rather than an actual missing data
