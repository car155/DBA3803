# Customising Plots

## Plot Styles and Palette

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

### Style

Style refers to the background and axes. There are a few preset options:
- white
- dark
- whitegrid
- darkgrid
- ticks

We can change the style using **sns.set_style** for all plots.

```python
tips = sns.load_dataset("tips")

sns.set_style("darkgrid")

sns.catplot(x="smoker", y="total_bill", data=tips, kind="point", hue="sex")
plt.show()
```

### Palette

Palette refers to the colors aof the main elements of the plot. Presets for it include:
- RdBu, RdBu_r
- PRGn, PRGn_r
- Greys, Blues, PuRd, GnBu

The palette of plots can be changed using the **sns.set_palette** function.

```python
tips = sns.load_dataset("tips")

custom_palette = ["#2A9D8F", "#E9C46A", "#F4A261", "#E76F51"]

sns.set_palette(custom_palette)

sns.catplot(x="day", y="total_bill", data=tips, kind="bar")
plt.show()
```

### Scale

Scale controls the size of the plot elements and labels. From smallest to largest, the options are:
- paper
- notebook
- talk
- poster

They can be set using the **sns.set_context** function.

```python
tips = sns.load_dataset("tips")

sns.set_context("talk")

sns.catplot(x="smoker", y="total_bill", data=tips, kind="point", hue="sex")
plt.show()
```

## Titles and Labels

Seaborn plots produce 2 different objects, a FacetGrid and an AxesSubplot

```python
tips = sns.load_dataset("tips")
g = sns.relplot(x="total_bill", y="tip", data=tips, col="sex")
type(g) # Facet grid
```

```python
height = [62, 64, 69, 75, 66, 68, 65, 71, 76, 73]
weight = [120, 136, 148, 175, 137, 165, 154, 172, 200, 187]
gender = ["Female", "Female", "Female", "Female", "Male", "Male", "Male", "Male", "Male", "Male"]

df = pd.DataFrame({"height":height, "weight":weight, "gender":gender})
df

g = sns.scatterplot(x="height", y="weight", data=df)
type(g) # Axes subplot
```

### Title

For facet grids:

```python
tips = sns.load_dataset("tips")
g = sns.relplot(x="total_bill", y="tip", data=tips, col="sex")

g.fig.suptitle("New Title", y=1.03)
plt.show()
```

For axes subplot:

```python
height = [62, 64, 69, 75, 66, 68, 65, 71, 76, 73]
weight = [120, 136, 148, 175, 137, 165, 154, 172, 200, 187]
gender = ["Female", "Female", "Female", "Female", "Male", "Male", "Male", "Male", "Male", "Male"]

df = pd.DataFrame({"height":height, "weight":weight, "gender":gender})
df

g = sns.scatterplot(x="height", y="weight", data=df)
g.set_title("New Title", y=1.03)
plt.show()
```

### Subplot Titles

```python
tips = sns.load_dataset("tips")
g = sns.relplot(x="total_bill", y="tip", data=tips, col="sex")

g.set_titles("Sex: {col_name}") # {col_name} allows us to use the category in the title
plt.show()
```

### Axes Labels

Setting labels:

```python
tips = sns.load_dataset("tips")
g = sns.relplot(x="total_bill", y="tip", data=tips, col="sex")

g.set(xlabel="New X Label", ylabel="New Y Label")
plt.show()
```

```python
height = [62, 64, 69, 75, 66, 68, 65, 71, 76, 73]
weight = [120, 136, 148, 175, 137, 165, 154, 172, 200, 187]
gender = ["Female", "Female", "Female", "Female", "Male", "Male", "Male", "Male", "Male", "Male"]

df = pd.DataFrame({"height":height, "weight":weight, "gender":gender})
df

g = sns.scatterplot(x="height", y="weight", data=df)
g.set(xlabel="New X Label", ylabel="New Y Label")
plt.show()
```

Rotating labels:

```python
tips = sns.load_dataset("tips")
g = sns.relplot(x="total_bill", y="tip", data=tips, col="sex")

for ax in g.axes.flat:
    for label in ax.get_xticklabels():
        label.set_rotation(90)
plt.show()
```

```python
height = [62, 64, 69, 75, 66, 68, 65, 71, 76, 73]
weight = [120, 136, 148, 175, 137, 165, 154, 172, 200, 187]
gender = ["Female", "Female", "Female", "Female", "Male", "Male", "Male", "Male", "Male", "Male"]

df = pd.DataFrame({"height":height, "weight":weight, "gender":gender})
df

g = sns.scatterplot(x="height", y="weight", data=df)
plt.xticks(rotation=90)
plt.show()
```
